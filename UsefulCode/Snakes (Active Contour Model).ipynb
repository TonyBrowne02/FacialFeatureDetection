```python
import numpy as np
from skimage import data
from skimage.filters import gaussian
from skimage.segmentation import active_contour
import matplotlib.pyplot as plt
```

```python
image = data.camera()
```

```python
# Initial Contour
s = np.linspace(0, 2*np.pi, 400)
x = 220 + 100*np.cos(s)
y = 100 + 100*np.sin(s)
init = np.array([x, y]).T
```

```python
# Apply gaussian to smooth the image
img_smt = gaussian(image, 3)

# Perform active contour
snake = active_contour(img_smt, init, alpha=0.015, beta=10, gamma=0.001)
```

```python
fig, ax = plt.subplots()
ax.imshow(image, cmap=plt.cm.gray)
ax.plot(init[:, 0], init[:, 1], '--r', lw=3)
ax.plot(snake[:, 0], snake[:, 1], '--b', lw=3)
plt.show()
```
